```python
def train(arch, item, batch, epochs=5, bs=64):
    dls = ImageDataLoaders.from_folder(trn_path, valid_pct=0.2, item_tfms=item, batch_tfms=batch, bs=bs)
    learn = vision_learner(dls, arch, metrics=error_rate).to_fp16()
    learn.fine_tune(epochs, 0.03)
    return learn
```

```python
learn = train('resnet26d', epochs=12, item=Resize(192), batch=aug_transforms(size=128, min_scale=0.75))
```

```python
# Full minscale, default p_affin - the best
learn = train('resnet26d', epochs=20, item=Resize(192), batch=aug_transforms(size=128, p_affine=.75))
```

```python
# arch = 'convnext_small_in22k' # Taking 5 mins to train.. too long
# arch = 'convnext_tiny_in22k' # Taking 3.5 mins to train.. Not bad, but still too long.
arch = 'resnet26d' # resnet 26 takees a minute to train with fairly good results.
```
