```python
from fastai.vision.all import *
```

```python
path = Path('.')
trn_path = path/'train_images'
im = PILImage.create((trn_path/'bacterial_leaf_blight'/'110320.jpg'))
print(im.size)
```

```python
480 / 640, 192 / 256
```

```python
dls = ImageDataLoaders.from_folder(trn_path, valid_pct=0.2, item_tfms=Resize((256,192)), bs=128)
# dls.show_batch()
```

```python
arch = resnet18
```

```python
learn = vision_learner(dls=dls, arch=arch, metrics=error_rate).to_fp16()
learn.lr_find()
```

```python
def train(arch, item, batch=aug_transforms(size=128), epochs=5, bs=64):
    dls = ImageDataLoaders.from_folder(trn_path, valid_pct=0.2, item_tfms=item, batch_tfms=batch, bs=bs)
    learn = vision_learner(dls, arch, metrics=error_rate).to_fp16()
    learn.fine_tune(epochs, 0.01)
    return learn
```

```python
learn = train(arch, epochs=20, item=Resize(224), bs=512)
```

```python
# item Resize into 
learn = train(arch, epochs=20, item=Resize(224))
```
